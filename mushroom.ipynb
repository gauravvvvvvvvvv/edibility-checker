```python
from sklearn.model_selection import train_test_split
x_train,  x_test,y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, chi2
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import OrdinalEncoder
```

```python
df.columns
```

```python
trf1 = ColumnTransformer(
    transformers=[
        ('ordinal', OrdinalEncoder(), ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor',
                                       'gill-attachment', 'gill-spacing', 'gill-size', 'gill-color',
                                       'stalk-shape', 'stalk-root', 'stalk-surface-above-ring',
                                       'stalk-surface-below-ring', 'stalk-color-above-ring',
                                       'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number',
                                       'ring-type', 'spore-print-color', 'population', 'habitat'])
    ],remainder='passthrough')
```

```python
trf2 =SelectKBest(score_func = chi2, k=8)
```

```python
trf3 = RandomForestClassifier()
#trf3 = LogisticRegression()
```

```python
pipe = pipeline = Pipeline([
    ("trf1", trf1),
    ("trf2", trf2),
    ("trf3", trf3),
])
```

```python
pipe.fit(x_train, y_train)
```

```python
#pipe analysis
```

```python
F = pipe.named_steps.trf2.get_support(indices = True)
```

```python
feat = x_train.columns[F]
```

```python
sns.barplot(x=feat, y = F)
plt.xticks(rotation=45)
plt.show()
```

```python
y_pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
